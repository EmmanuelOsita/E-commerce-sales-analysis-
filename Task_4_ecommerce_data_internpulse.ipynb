**Objective 2: Customer Retention and Loyalty**

Calculate the customer retention rate and identify the percentage of returning customers.


```python
# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Count the number of orders per customer
customer_order_count = df.groupby('customer_id')['order_id'].nunique().reset_index()
customer_order_count.columns = ['customer_id', 'order_count']

# Identify returning customers (order_count > 1)
returning_customers = customer_order_count[customer_order_count['order_count'] > 1]

# Calculate retention rate
total_customers = customer_order_count['customer_id'].nunique()
returning_customer_count = returning_customers['customer_id'].nunique()
retention_rate = (returning_customer_count / total_customers) * 100
```

```python
# Output results
print(f"Total Unique Customers: {total_customers}")
print(f"Returning Customers: {returning_customer_count}")
print(f"Customer Retention Rate: {retention_rate:.2f}%")
```

**Objective 4 Delivery and Sales Relationship**

 Assess how delivery status (e.g., on-time, delayed, canceled) impacts sales and customer satisfaction in various regions.


```python
# 1. Aggregate sales and orders by delivery status
delivery_sales = df.groupby("delivery_status").agg(
    total_sales=("sales_per_order", "sum"),
    average_sales=("sales_per_order", "mean"),
    total_orders=("order_quantity", "sum"),
).reset_index()

# 2. Aggregate sales by delivery status and region
regional_delivery_sales = df.groupby(["customer_region", "delivery_status"]).agg(
    total_sales=("sales_per_order", "sum"),
    total_orders=("order_quantity", "sum")
).reset_index()

# 3. Visualize sales by delivery status
plt.figure(figsize=(8, 6))
sns.barplot(data=delivery_sales, x="delivery_status", y="total_sales", palette="Set2")
plt.title("Total Sales by Delivery Status")
plt.xlabel("Delivery Status")
plt.ylabel("Total Sales")
plt.show()

# 4. Visualize regional impact of delivery status on sales
plt.figure(figsize=(10, 6))
sns.barplot(data=regional_delivery_sales, x="customer_region", y="total_sales", hue="delivery_status", palette="muted")
plt.title("Regional Sales by Delivery Status")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.legend(title="Delivery Status")
plt.show()

# Print tables for reference
print("Sales by Delivery Status:\n", delivery_sales)
print("\nRegional Sales by Delivery Status:\n", regional_delivery_sales)
```

**Objective 5. Shipping Type vs. Delivery Performance**

 Analyze how different shipping options (e.g., standard, expedited, or same-day) influence delivery performance.


```python
# 1. Group data by shipping type and delivery status
shipping_performance = df.groupby(["shipping_type", "delivery_status"]).size().unstack(fill_value=0)

# 2. Calculate proportions for each shipping type
shipping_proportions = shipping_performance.div(shipping_performance.sum(axis=1), axis=0) * 100

# 3. Visualize the data
shipping_proportions.plot(kind="bar", stacked=True, figsize=(10, 6), colormap="viridis")
plt.title("Delivery Performance by Shipping Type")
plt.xlabel("Shipping Type")
plt.ylabel("Percentage of Deliveries")
plt.legend(title="Delivery Status")
plt.show()

# Print tables for reference
print("Shipping Performance (Counts):\n", shipping_performance)
print("\nShipping Performance (Proportions):\n", shipping_proportions)
```

**Objective 6. Customer Segmentation**

 a. Perform customer segmentation to identify groups based on preferences, behavior, or demographics.

 b. Discover which segments show the most interest in specific products or categories.


```python
# Step 1: Aggregate data to find total spending, quantity, and profits per customer
customer_data = data.groupby('customer_id').agg({
    'sales_per_order': 'sum',
    'order_quantity': 'sum',
    'profit_per_order': 'sum',
    'category_name': lambda x: x.mode()[0]

# Most common category for each customer
}).reset_index()

# Normalize the data for clustering
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
customer_features = scaler.fit_transform(customer_data[['sales_per_order', 'order_quantity', 'profit_per_order']])

# Step 2: Apply K-Means Clustering
#Import KMeans
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['Segment'] = kmeans.fit_predict(customer_features)

# Step 3: Analyze preferences for each segment
segment_analysis = customer_data.groupby('Segment').agg({
    'sales_per_order': 'mean',
    'order_quantity': 'mean',
    'profit_per_order': 'mean',
    'category_name': lambda x: x.mode()[0]  # Most preferred category
}).reset_index()

# Step 4: Visualize the segments
plt.figure(figsize=(10, 6))
sns.barplot(data=segment_analysis, x='Segment', y='sales_per_order', palette='viridis')
plt.title('Average Sales per Segment')
plt.xlabel('Segment')
plt.ylabel('Average Sales')
plt.show()
```
